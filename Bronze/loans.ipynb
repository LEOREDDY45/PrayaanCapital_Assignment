```python
df_schema=StructType([
  StructField('loan_id', StringType(), True),
  StructField('branch_id', StringType(), True),
  StructField('borrower_id', StringType(), True),
  StructField('disbursal_date', DateType(), True),
  StructField('principal', IntegerType(), True),
  StructField('tenure_months', IntegerType(), True),
  StructField('interest_rate_annual_pct', FloatType(), True)
])
```

```python
df=spark.read.schema(df_schema).csv(bronze+'/loans.csv', header=True)
```

```python
df=df.filter((col("disbursal_date").isNotNull()) & 
             (col('disbursal_date')<=current_date()) & 
             (col("principal")>0) & (col("tenure_months")>0) & 
             (col("interest_rate_annual_pct")>0))
df=df.filter( col("borrower_id").isNotNull() | col("branch_id").isNotNull() |col("borrower_id").isNotNull())
```

```python
df.write.mode('overwrite').csv(silver+'/loans.csv', header=True)
```
