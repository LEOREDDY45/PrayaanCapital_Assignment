```python
df_schema=StructType(       [StructField('repayment_id', StringType(), True),
                             StructField('loan_id', StringType(), True),
                             StructField('installment_no', IntegerType(), True),
                             StructField('due_date', DateType(), True),
                             StructField('paid_date', DateType(), True),
                             StructField('amount_due', FloatType(), True),
                             StructField('amount_paid', FloatType(), True)])
```

```python
df=spark.read.option('header', True).schema(df_schema).csv(bronze+'/repayments.csv')
```

```python
df=df.filter((col('paid_date').isNotNull()) 
             & (col("paid_date") <= current_date()) 
             & (col("amount_paid") > 0) 
             & (col("amount_due") > 0))
```

```python
df.write.mode('overwrite').option('header', True).format('csv').save(silver+'/repayments.csv')
```
